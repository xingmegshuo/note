```python
from keras.layers import Convolution2D,Activation,MaxPooling2D
from keras.models import Sequential
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
picture = Image.open('a.jpg')
plt.imshow(picture)
```

```python
picture = np.array(picture)
print(picture.shape)
model = Sequential()
model.add(Convolution2D(3,18,18,input_shape=picture.shape))
model.add(Activation('relu'))
picture_batch = np.expand_dims(picture,axis=0)
print(picture_batch.shape)
# print(picture_batch.shape)
conv_picture = model.predict(picture_batch)
conv_pic = np.squeeze(conv_picture,axis=0)
print(conv_pic.shape)
# conv_pic = conv_pic.reshape(conv_pic.shape[:2])
plt.imshow(conv_pic)
```

```python
tree = Image.open('tree.jpeg')
plt.imshow(tree)
tree = np.array(tree)
print(tree.shape)
```

```python
model = Sequential()
model.add(Convolution2D(1,5,5,input_shape=tree.shape))
model.add(Activation('relu'))
model.add(Convolution2D(1,3,3))
model.add(Activation('relu'))
picture_batch = np.expand_dims(tree,axis=0)
conv_tree = model.predict(picture_batch)
conv_tree = np.squeeze(conv_tree,axis=0)
print(conv_tree.shape)
conv_pic = conv_tree.reshape(conv_tree.shape[:2])
plt.imshow(conv_pic)
```

```python
#
cat = Image.open('cat.jpeg')
plt.imshow(cat)
cat = np.array(cat)
```

```python
model = Sequential()
model.add(Convolution2D(1,3,3,input_shape = cat.shape))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(3,3)))
cat_batch = np.expand_dims(cat,axis=0)
conv_cat = model.predict(cat_batch)
conv_cat = np.squeeze(conv_cat,axis=0)
conv_pic = conv_cat.reshape(conv_cat.shape[:2])
plt.imshow(conv_pic)
```

```python
def commpression(string):
    new_str=''
    count=0
    temp=string[0]
    for i in range(0,len(string)):
        if temp==string[i]:
            count+=1
        else:
            new_str=new_str+str(temp)+str(count)
            count=1
            temp=string[i]
        if i == len(string)-1:
            new_str=new_str+str(temp)+str(count)  
    return new_str

print(commpression('aaaaacccccdddddddbbbbbqqqq'))
```

```python
from itertools import groupby
```

```python
def decompression(string):
    new_str=''
    string_list = [''.join(list(k)) for g,k in groupby(string,key=lambda x:x.isalpha())]
    for i in range(0,len(string_list)):
        if i%2 == 1:
            new_str=new_str+string_list[i-1]*int(string_list[i])
    return new_str
        
    
```

```python
decompression('a10b26c7e6')
```

```python
import numpy as np

info = np.array([[1,3,9],
                 [14,15,20]])
info_A = np.mat(info)   #使用 mat 方法将 2 维数组转化为矩阵

key = np.array([[3,5,7],
                [4,7,20],
                [100,0,0]])
key_A = np.mat(key)   #使用 mat 方法将 2 维数组转化为矩阵
print("密钥：",key_A)
after_A = info_A*key_A
print("加密后：",after_A)
print("#######################")
print("解密")
key1 = np.array([[3,5,7],
                [4,7,20],
                [100,0,0]])
key1_A = np.mat(key1)   #使用 mat 方法将 2 维数组转化为矩阵
print(after_A*key1_A.I)
```

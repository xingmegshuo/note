```python
# 两个分类，生成两个重心
k = 2
k1=[random.uniform(1,2),random.uniform(1,2)]
k2 = [random.uniform(2,3),random.uniform(1,2)]
#计算距离
def computer_distance(A,B):
    distance = np.sqrt(np.square(A[0]-B[0])+np.square(A[1]-B[1]))
    return distance
# 计算重心
def compute_weightcenter(l):
    a=0
    b=0
    for i in l:
        a+=i[0]
        b+=i[1]
    return [a/len(l),b/len(l)]
        
    
        
```

```python
a=10
count=0
while count < a:
    print(k1,k2)
#     plot_data()
    plt.scatter(k1[0],k1[1],c='b',marker='+')
    plt.scatter(k2[0],k2[1],c='r',marker='+')
    k1_cu=[]
    k2_cu=[]
    for i in range(data.shape[1]):
        k1_distance = computer_distance(data[:,i],k1)
        k2_distance = computer_distance(data[:,i],k2)
        if k1_distance < k2_distance:
            k1_cu.append(data[:,i])
        else:
            k2_cu.append(data[:,i])
    l1 = [i[0] for i in k1_cu]
    l2 = [i[1] for i in k1_cu]
    l3 = [i[0] for i in k2_cu]
    l4 = [i[1] for i in k2_cu]
    plt.scatter(l1,l2,c='r')
    plt.scatter(l3,l4,c='b')
    plt.show()
    k1 = compute_weightcenter(k1_cu)
    k2 = compute_weightcenter(k2_cu)
    count += 1
```
